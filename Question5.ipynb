### Predicted vs Expected results

```python
from sklearn.metrics import mean_squared_error
X = data.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```

###  Choice for Autoregressive Moving Average (ARMA) or Autoregressive Moving Average (ARIMA) method was made because it is suitable for univariate time series without trend and seasonal components, which is what our data is. 
### It combines both Auto regression and Moving Average models and predicts better results then just using AR or MA.
### I have used ARIMA to take out any trend from our data, differncing the data only once.
